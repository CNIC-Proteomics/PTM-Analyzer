```python
import pandas as pd
import os
import numpy as np

idx = pd.IndexSlice
```

```python
df = pd.read_csv(
    r"S:\U_Proteomica\LABS\LAB_DF\Comet_PTM\FA_Biopsias\2_ReportLimma_wo_GUI_oneSample\NM_Tabla_final.tsv",
    sep='\t', header=[0,1], low_memory=False
)
```

```python
# Set groups

groups = {
    'B0': ['B0_A','B0_B','B0_C','B0_D','B0_E','B0_F'],
    'F': ['F_A','F_B','F_C','F_D','F_E','F_F'],
    'I': ['I_A','I_B','I_C','I_D','I_E','I_F']
}
```

```python
list(set([i for i,j in df.columns if j=='LEVEL']))
df#['Z_pgm2p_dNM']
```

```python
integrations = [('Z_pdm2pgm', 'pdm'), ('Z_pgm2p', 'pgm'), ('Z_pgm2p_dNM', 'pgm'),
                ('Z_p2qf', 'p'), ('Z_qf2q', 'qf'), ('Z_q2all', 'q')]

intgr, lowL = integrations[0]

for g in groups:

    if g not in os.listdir():
        os.mkdir(g)  

    for intgr, lowL in integrations:

        if intgr not in os.listdir(g):
            os.mkdir(os.path.join(g, intgr))

        df.loc[:, [(lowL, 'LEVEL')]].join(
            df.loc[:, idx[intgr, :]].dropna(axis=0, how='all'),
            how='right'
        ).drop_duplicates().droplevel(0, axis=1).loc[:, ['LEVEL']+groups[g]].to_csv(os.path.join(g, intgr, 'input.tsv'), sep='\t', index=False)
```

After executing LIMMA...

```python
intgr, lowL = integrations[0]

df_out = df.copy()

#intgr, lowL = integrations[0]
for g in groups:
    for intgr, lowL in integrations:

        limma_output = pd.read_csv(os.path.join(g, intgr, 'output.tsv'), sep='\t', low_memory=False)
        limma_output.columns = pd.MultiIndex.from_tuples([(lowL, 'LEVEL'), (f'{intgr}_dX_{g}', 'dX'), (f'{intgr}_limma_{g}', 'pvalue')])
        limma_output[f'{intgr}_logLimma_{g}', 'LPS'] = -np.log10(limma_output[(f'{intgr}_limma_{g}', 'pvalue')])*np.sign(limma_output[(f'{intgr}_dX_{g}', 'dX')])

        df_out = pd.merge(
            df_out, limma_output,
            how='left',
            on=[(lowL, 'LEVEL')]
            )
```
